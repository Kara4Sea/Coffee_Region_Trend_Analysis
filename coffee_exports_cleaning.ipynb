```python
# import dependencies
import pathlib as path
import numpy as np
import pandas as pd
```

```python
# export csv into dataframe
df_export = pd.read_csv("Resources/coffee_exports.csv")
df_export
```

```python
#find dtypes
df_export.dtypes
```

```python
#find sum duplicated country values
df_export.country.duplicated().sum()
```

```python
# drop columns to reflect most recent 10 years. And remove all_years column.
df_export2 = df_export.drop(columns=['1990', '1991', '1992', '1993', '1994', '1995', '1996', '1997','1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', 'all_years'])
df_export2.head()
```

```python
# convert to kg count (originally in thousands of 60kg bags)
df_export3 = df_export2.select_dtypes(exclude=['object']) * 60000
df_export3.head()
```

```python
#convert from kg to lbs
df_export4 = df_export3 * 2.20462
df_export3.head()
```

```python
#show in millions lbs
df_export5 = df_export4 / 1000000
df_export5.head()
```

```python
# Assign country names to new df
df_export6 = df_export2.drop(columns=['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017','2018', '2019'])
df_export6.head()
```

```python
# combine country name and counts by millions lbs into one df
df_exports_by_country = pd.concat([df_export6, df_export5], axis=1, join='inner')
df_exports_by_country
```

```python
# Create new all_years column for 10 year time span
col = df_exports_by_country.loc[: , "2010":"2019"]
df_exports_by_country['all_years'] = col.sum(axis=1)
df_exports_by_country.head()
```

```python
df_exports_by_country_volume = df_exports_by_country.sort_values(by=['all_years'], ascending=False)
df_exports_by_country_volume.head(11)
```

```python
# limit to top ten countries by all-years volume
df_exports_by_country_top_ten = df_exports_by_country_volume.drop(55)
df_exports_by_country_top_ten = df_exports_by_country_top_ten.head(10)
df_exports_by_country_top_ten
```

```python
file_path = "Resources/coffee_exports_top_ten_lbs.csv"
df_exports_by_country_top_ten.to_csv(file_path, index=False)
```
